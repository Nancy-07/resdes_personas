```python
import numpy as np
import pandas as pd


# Leer el dataset desde un archivo CSV
df = pd.read_csv("resultados_model_primera100.csv", header=None)

# Obtener características (X) y etiquetas (y) del DataFrame
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
import matplotlib.pyplot as plt

plt.plot(y,linestyle='--', color='r', label='Datos')
plt.xlabel('Índice')
plt.ylabel('Presición')
plt.title('Presición con arquitectura de una capa')
plt.legend()
print(max(y))
```
